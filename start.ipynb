```python
#sample images

import matplotlib.pyplot as plt
from PIL import Image
import random

def show_samples(folder, label):
    images = os.listdir(folder)
    sample_images = random.sample(images, 5)

    plt.figure(figsize=(10,4))
    for i, img_name in enumerate(sample_images):
        img = Image.open(os.path.join(folder, img_name))
        plt.subplot(1,5,i+1)
        plt.imshow(img)
        plt.axis("off")
        plt.title(label)

show_samples("../data/classification_dataset/TRAIN/bird", "Bird")
show_samples("../data/classification_dataset/TRAIN/drone", "Drone")
```

```python
#checking dimensions of images

from collections import Counter

sizes = []

for cls in ["bird", "drone"]:
    folder = f"../data/classification_dataset/TRAIN/{cls}"
    for img_name in os.listdir(folder)[:50]:
        img = Image.open(os.path.join(folder, img_name))
        sizes.append(img.size)

print(Counter(sizes).most_common(5))
```
